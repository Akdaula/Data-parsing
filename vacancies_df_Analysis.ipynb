```python
df['Highest Salary'] = df['Salary'].str.extract(r'(\d+)\s*(?:[A-Z]{3}|None)$')[0]

df
```

```python
exchange_rates = {'RUR': 0.0105, 'USD': 1, 'EUR': 1.05, 'KZT': 0.0021}


extracted_data = df['Salary'].str.extract(r'(\d+)\s*([A-Z]{3}|None)$')
df['Highest Salary Amount'] = extracted_data[0].astype(float)  
df['Currency'] = extracted_data[1].replace('None', 'USD')  

df['Salary in USD'] = df.apply(lambda row: row['Highest Salary Amount'] * exchange_rates.get(row['Currency'], 1), axis=1)

print(df[['Salary', 'Salary in USD']])
```

```python
df
```

```python
df['Salary in USD'].isnull().sum()
```

```python
# Step 1: Calculate the median salary by City ID
median_salary_by_city = df.groupby('City ID')['Salary in USD'].median()

# Step 2: Fill missing values (NaN) in 'Salary in USD' with the median salary for the corresponding City ID
df['Salary in USD'] = df.apply(
    lambda row: median_salary_by_city[row['City ID']] if pd.isna(row['Salary in USD']) else row['Salary in USD'], axis=1
)


```

```python
city_mapping = {
    1: 'Moscow',
    2:'St.Peterburg',
    159: 'Almaty',
    160:'Astana'
}

df['City'] = df['City ID'].map(city_mapping)
```

# EDA 

## Hypotheses 

- In the cities of Almaty and Astana salaries are significantly lower compared to Moscow and St. Petersburg
- There are many more vacancies in the regions of Russia than in the cities of Kazakhstan

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['City'], df['Salary in USD'], color=['blue'])
plt.title('Average Salaries in Almaty, Astana, Moscow, and St. Petersburg')
plt.xlabel('City')
plt.ylabel('Average Salary in USD')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
df.boxplot(by='City', column=['Salary in USD'], grid=True, patch_artist=True)
plt.title('Salary Distribution in Moscow, St. Petersburg, Astana, and Almaty')
plt.suptitle('')  
plt.xlabel('City')
plt.ylabel('Salary in USD')
plt.show()
```

```python
city_counts = df['City'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue','lightblue', 'lightgreen','lightgreen'])
plt.title('Distribution of Cities in the Dataset')
plt.show()
```
